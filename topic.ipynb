```python
from transformers import pipeline
import torch

# Use GPU if available
device = 0 if torch.cuda.is_available() else -1

# Load GPT-2 text generator
generator = pipeline("text-generation", model="gpt2", device=device)
print("✅ GPT-2 ready")
```

```python
prompt = "The future of artificial intelligence in healthcare is"
result = generator(prompt, max_length=150, temperature=0.8, top_k=40, top_p=0.9, do_sample=True)[0]['generated_text']

print("=== Generated Paragraph ===\n")
print(result)
```

```python
user_prompt = input("Enter your topic: ")
result = generator(user_prompt, max_length=150, temperature=0.8, top_k=40, top_p=0.9, do_sample=True)[0]['generated_text']

print("\n=== Generated Paragraph ===\n")
print(result)
```
